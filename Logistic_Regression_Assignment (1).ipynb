#Encoding Categorical Data

```python
# One-Hot Encoding of categrical variables
d2=pd.get_dummies(dataset,columns=['job','marital','education','contact','poutcome'])
d2
```

```python
d2.info()
```

```python
# Custom Binary Encoding of Binary o/p variables 
d2['default'] = np.where(d2['default'].str.contains("yes"), 1, 0)
d2['housing'] = np.where(d2['housing'].str.contains("yes"), 1, 0)
d2['loan'] = np.where(d2['loan'].str.contains("yes"), 1, 0)
d2['y'] = np.where(d2['y'].str.contains("yes"), 1, 0)
d2
```

```python
d2.info()
```

```python
d2['month'].value_counts()
```

```python
#give no. to the month name.
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
d2=d2.replace(order)
```

```python
d2
```

```python
d2.info()
```

# Independent and Dependent Variables

```python
# Dividing our data into input and output variables
X=pd.concat([d2.iloc[:,0:11],d2.iloc[:,12:]],axis=1)
y=d2.iloc[:,11]
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
X
```

```python
y
```

# Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
y_train
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
X_test
```

# MACHINE LEARNING ALGORITHM

## LOGISTIC REGRESSION

```python
#Fitting the model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train,y_train)
```

```python
#Predict my test results
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

### Model is 89% Accurate

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

We got an accuracy score of around 90% on the test data
